```python
import torch
import matplotlib.pyplot as plt
from torchvision import transforms, models, datasets
import numpy as np
import json

mean = np.array([0.485, 0.456, 0.406], dtype=np.float32)
std  = np.array([0.229, 0.224, 0.225], dtype=np.float32)
inv_norm = transforms.Normalize(mean=(-mean / std).tolist(), std=(1 / std).tolist())

transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean.tolist(), std.tolist())])
dataset = datasets.ImageFolder(root="TestDataSet", transform=transform)
loader = torch.utils.data.DataLoader(dataset, batch_size=64, shuffle=False)

with open("TestDataSet/labels_list.json") as f:
    idx2true = {i: int(e.split(":",1)[0]) for i, e in enumerate(json.load(f))}

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = models.resnet34(weights=models.ResNet34_Weights.IMAGENET1K_V1).to(device)
model.eval()
```

```python
adv_50 = torch.load("adv_set_patch_soft_eps0.3_step50.pt")
adv_80 = torch.load("adv_set_patch_soft_eps0.3_step80.pt")
adv_110 = torch.load("adv_set_patch_soft_eps0.3_step110.pt")

adv_imgs_50 = adv_50["images"]
adv_imgs_80 = adv_80["images"]
adv_imgs_110 = adv_110["images"]
labels = adv_50["labels"]

orig_images = []
for x, _ in loader:
    orig_images.append(x)
    if len(orig_images) * 64 >= len(labels):
        break
orig_images = torch.cat(orig_images)[:len(labels)].to(device)
```

```python
@torch.no_grad()
def get_predictions(model, tensor, batch_size=64):
    preds = []
    for i in range(0, len(tensor), batch_size):
        batch = tensor[i:i+batch_size].to(device)
        logits = model(batch)
        preds.extend(logits.argmax(1).cpu())
    return torch.tensor(preds)

clean_preds = get_predictions(model, orig_images)
eps50_preds = get_predictions(model, adv_imgs_50)
eps80_preds = get_predictions(model, adv_imgs_80)
eps110_preds = get_predictions(model, adv_imgs_110)
true_labels = torch.tensor([idx2true[int(l)] for l in labels])
```

```python
all_labels = models.ResNet34_Weights.IMAGENET1K_V1.meta["categories"]

picked = []
for i, (c, p50, p80, p110, t) in enumerate(zip(clean_preds, eps50_preds, eps80_preds, eps110_preds, true_labels)):
    if c == t and (p50 != t or p80 != t or p110 != t):
        picked.append(i)
    if len(picked) >= 5:
        break

fig, axes = plt.subplots(4, len(picked), figsize=(6, 6))
names = ["Clean", "Step=50", "Step=80", "Step=110"]
image_sets = [orig_images, adv_imgs_50, adv_imgs_80, adv_imgs_110]
pred_sets  = [clean_preds, eps50_preds, eps80_preds, eps110_preds]

for row in range(4):
    for col, i in enumerate(picked):
        img = inv_norm(image_sets[row][i].cpu()).permute(1, 2, 0).clamp(0, 1)
        pred_label = all_labels[pred_sets[row][i]]
        axes[row, col].imshow(img)
        axes[row, col].axis("off")
        axes[row, col].set_title(f"{names[row]}\n{pred_label}", fontsize=8)

plt.suptitle("Clean vs PGD Patch =0.3 with Different Steps", fontsize=14)
plt.tight_layout()
plt.show()
```
